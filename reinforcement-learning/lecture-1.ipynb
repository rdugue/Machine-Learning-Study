# CS234 - Lecture 1-3 Q-Tables

```python
import gymnasium as gym
import numpy as np

# Create the environment (e.g., 'FrozenLake')
env = gym.make("FrozenLake-v1")
total_episodes = 100
Q = np.zeros((16, 4))
epsilon = 1.0
epsilon_min = 0.01
epsilon_decay = 0.995

# The Core Q-Learning Loop
for episode in range(total_episodes):
    state, info = env.reset()
    terminated = False
    truncated = False
    gamma = 1.0
    alpha = 0.1
    
    while not terminated and not truncated:
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample() # Explore
        else:
            action = np.argmax(Q[state])       # Exploit
        # 2. Step
        new_state, reward, terminated, truncated, info = env.step(action)
        
        # 3. Calculate Target
        target = reward if terminated else reward + gamma * np.max(Q[new_state])
        
        # 4. Update Q-Table
        Q[state, action] = Q[state, action] + alpha * (target - Q[state, action])
        
        # 5. Move State
        state = new_state

    # Decay epsilon, but don't let it drop below a minimum value
    epsilon = max(epsilon_min, epsilon * epsilon_decay)
```
